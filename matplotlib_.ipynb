# **Ploting with bar plot**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# X parametrr
x = ["C", "C++", "Python", "Java", "javascript"]
y = [ 60, 70, 85, 75, 90]
```

```python
#ploting as bar ploting
plt.figure(figsize=(4, 3))
plt.bar(x, y)
plt.xlabel("Programing Languages", fontsize=10)
plt.ylabel("Popularrity in %", fontsize= 10)
plt.title("Day by day increasing or decreasing demand of programming language", fontsize= 10)
plt.show()
```

```python
#decreasing width of bars
plt.figure(figsize=(4, 3))
plt.bar(x, y, width=0.4)
plt.xlabel("Programing Languages", fontsize=10)
plt.ylabel("Popularrity in %", fontsize= 10)
plt.title("Day by day increasing or decreasing demand of programming language", fontsize= 10)
plt.show()
```

```python
# changing colors of Bar
c = ["r", "b", "m", "r", "g"]
plt.figure(figsize=(4, 3))
plt.bar(x, y, width=0.4, color=c)
plt.xlabel("Programing Languages", fontsize=10)
plt.ylabel("Popularrity in %", fontsize= 10)
plt.title("Day by day increasing or decreasing demand of programming language", fontsize= 10)
plt.show()
```

```python
# changing line width and style of bars
plt.figure(figsize=(4, 3))
plt.xlabel("Programing Languages", fontsize=10)
plt.ylabel("Popularrity in %", fontsize= 10)
plt.title("Day by day increasing or decreasing demand of programming language", fontsize= 10)
plt.bar(x, y, width=0.4, color=c, edgecolor="r", linewidth=5, linestyle=":")
plt.show()
```

```python
#using label on plot area
plt.figure(figsize=(4, 3))
plt.xlabel("Programing Languages", fontsize=10)
plt.ylabel("Popularrity in %", fontsize= 10)
plt.title("Day by day increasing or decreasing demand of programming language", fontsize= 10)
plt.bar(x, y, width=0.4, color=c, label="Popularity")
plt.legend()
plt.show()
```

```python
#ploting multile bar for one element

z=[20, 30, 40, 50, 80]
cz = ["b", "r", "r", "y", "r"]
p = np.arange(len(x))
width = 0.4
p1 = [j+width for j in p]
plt.figure(figsize=(4, 3))
plt.xlabel("Programing Languages", fontsize=10) # naming fro x label
plt.ylabel("Popularrity in %", fontsize= 10)
plt.title("Day by day increasing or decreasing demand of programming language", fontsize= 10)
plt.bar(x, y, width=0.4, color=c, label="Popularity")
plt.bar(p1 , z, width=0.4, color=cz, label="Popularity1")
plt.xticks(p+width/2, x, rotation=20) # rotation for tild bar labels
plt.legend() # use lagend() for showing title on th plot area
plt.show()
```

```python
#creating horizontal bar graph
plt.figure(figsize=(4, 3))
plt.xlabel("Programing Languages", fontsize=10)
plt.ylabel("Popularrity in %", fontsize= 10)
plt.title("Day by day increasing or decreasing demand of programming language", fontsize= 10)
plt.barh(x, y, color =c, label="Ploularity" )
plt.legend()
plt.show()
```

```python
# adding title and xlabel and y label
day = np.arange(1, 15)
no = np.array([2, 1, 4, 6, 3, 4, 8, 7, 9, 12, 13, 3, 14, 11])
```

```python
print(len(day))
print(len(no))
```

```python
#coloring the dots
sc = np.arange(1, 15)
plt.title("Day number plot", fontsize=10)
plt.xlabel("Day", fontsize=10)
plt.ylabel("No.", fontsize=10)
plt.scatter(day , no, c=sc)
plt.show()
```

```python
# resizing scater dots
sc = np.arange(1,15)
size = [j**3 for j in range(1, 15)]
plt.title("Day number plot", fontsize=10)
plt.xlabel("Day", fontsize=10)
plt.ylabel("No.", fontsize=10)
plt.scatter(day , no, c=sc , s= size)
plt.show()
```

```python
#Changing color using color map(cmap) and color bar
sc = np.arange(1,15)
size = [j**3 for j in range(1, 15)]
plt.title("Day number plot", fontsize=10)
plt.xlabel("Day", fontsize=10)
plt.ylabel("No.", fontsize=10)
plt.scatter(day , no, c=sc , s= size, cmap='Accent')
plt.show()
```

```python
sc = np.arange(1,15)
size = [j**3 for j in range(1, 15)]
plt.title("Day number plot", fontsize=10)
plt.xlabel("Day", fontsize=10)
plt.ylabel("No.", fontsize=10)
plt.scatter(day , no, c=sc , s= size, cmap='viridis')
plt.show()
```

```python
#show color bar
sc = np.arange(1,15)
size = [j**3 for j in range(1, 15)]
plt.title("Day number plot", fontsize=10)
plt.xlabel("Day", fontsize=10)
plt.ylabel("No.", fontsize=10)
plt.scatter(day , no, c=sc , s= size, cmap='CMRmap')
clbar = plt.colorbar()
clbar.set_label("Color bar", fontsize=15)
plt.show()
```

```python
#Adding Label to color bar
sc = np.arange(1,15)
size = [j**3 for j in range(1, 15)]
plt.title("Day number plot", fontsize=10)
plt.xlabel("Day", fontsize=10)
plt.ylabel("No.", fontsize=10)
plt.scatter(day , no, c=sc , s= size, cmap='PiYG')
clbar = plt.colorbar()
clbar.set_label("Color bar", fontsize=15)
plt.show()
```
